```python
import pandas as pd
from parsers.cpu_parser import extract_info

path = "/Users/mohammad/Documents/Python/projects/sepidlaptop/input_data/goldkala_laptops.csv"

rawdata = pd.read_csv(path)

pd.set_option("display.max_rows", None)

cleandf = pd.DataFrame()


def model_ext():
    
    pattern= r"(\w[\u0600-\u06FF]+)"
    
    brand= list(rawdata["برند"].str.lower().str.strip())
    
    cleandf["model"] = rawdata["مشخصات لپ تاپ"].str.split("|").str[0].str.replace(pattern,"",regex=True).str.lower()
    
    for b in brand:
        
        pattern = rf"({b})"
        
        cleandf["model"] = cleandf["model"].str.replace(pattern,"", regex=True).str.strip()
    
    
def brand_ext():
    
    cleandf["brand"] = rawdata["برند"].str.strip().str.lower()
    

def cpu_ext():
    
    cpu_df = rawdata["پردازنده"].apply(extract_info).apply(pd.Series)
    
    cleandf[cpu_df.columns] = cpu_df
    
    
    


def main():
    
    brand_ext()
    model_ext()
    cpu_ext()
    
    
    
main()
cleandf

```
